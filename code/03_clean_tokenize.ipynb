### Loading in initial data, setting jupyter defaults, and downloading assets

```python
import pandas as pd
import re
import csv
from IPython.display import display, clear_output
import nltk 
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.tokenize.treebank import TreebankWordDetokenizer
import string
from string import capwords
from sklearn.feature_extraction.text import CountVectorizer
import geograpy3
from cliff.api import Cliff
import json 
```

```python
df = pd.concat(map(pd.read_csv, ['../data/abstracts_chunk_1.csv', '../data/abstracts_chunk_2.csv', 
                                    '../data/abstracts_chunk_3.csv','../data/abstracts_chunk_4.csv','../data/abstracts_chunk_5.csv',
                                  '../data/abstracts_chunk_6.csv','../data/abstracts_chunk_7.csv', '../data/abstracts_chunk_8.csv',
                                    '../data/abstracts_chunk_11.csv']), ignore_index=True)
```

```python
# Display settings
pd.set_option('display.max_columns', 999)
pd.set_option('display.max_rows', 999)

# drop rows with nan abstracts
df = df.dropna(subset=['abstracts-retrieval-response.coredata.dc:description.abstract.ce:para'])
print(df['abstracts-retrieval-response.coredata.dc:description.abstract.ce:para'].isnull().sum())

# download stopwords
nltk.download('stopwords')
nltk.download('punkt')

def getTdm(doc):
    vec = CountVectorizer()
    x = vec.fit_transform(doc)
    df = pd.DataFrame(x.toarray(), columns=vec.get_feature_names())
    return df
```

### Remove unecessary index column 

```python
if 'Unnamed: 0' in df:
    del  df['Unnamed: 0']
```

```python
abstract_df = pd.DataFrame()
abstract_df['original_text'] = df['abstracts-retrieval-response.coredata.dc:title'] + df['abstracts-retrieval-response.coredata.dc:description.abstract.ce:para']
abstract_df['text'] = abstract_df['original_text'].str.lower()
display(abstract_df.head())
```

### Removing nonalphabetic strings and stopwords

```python
# Initializing stopwords and regex strings
stop_words = set(stopwords.words('english')) 
non_alphabetical = re.compile("[^A-Z^a-z^\s]")

# Adding new data columns to df
abstract_df['text'] = abstract_df['text'].astype(str).replace(non_alphabetical, "")
abstract_df['original_text'] = abstract_df['original_text'].astype(str).replace(non_alphabetical, "")
abstract_df['tokenized_text'] = abstract_df['text'].apply(nltk.word_tokenize)
abstract_df['no_stopwords_tokenized_text'] = abstract_df['tokenized_text'].apply(lambda x: [word for word in x if word not in stop_words])
abstract_df['no_stopwords_text'] = abstract_df['no_stopwords_tokenized_text'].apply(lambda x: ' '.join(x))
```

```python
# Visualization of abstract
display(abstract_df.head())
print(len(df.index))
```

```python
# Term Document Matrix Creation
abstract_df['tdm'] = abstract_df['no_stopwords_text'].apply(lambda x: getTdm([x]))
```
