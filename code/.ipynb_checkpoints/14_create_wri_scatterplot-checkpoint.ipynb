```python
def format_element(element):
    country_reformat = {
#         'america': 'united states',
#         'united states of america':'united states',
#         'USA': 'united states',
#         'congo' : 'congo, dem. rep.',
#         'egypt' : 'egypt, arab rep.',
#         'yemen' : 'yemen, rep.',
#         'bahamas' : 'bahamas, the',
#         'russia' : 'russian federation',
#         'united kingdom of great britain and northern ireland': 'united kingdom'
        'republic of congo' :'congo, rep.',
        'democratic republic of the congo':'congo, dem. rep.',
        'laos':'lao pdr',
        'slovakia':'slovak republic',
        'brunei':'brunei darussalam',
        'kyrgyzstan': 'kyrgyz republic',
        'south korea':'korea, rep.',
       "north korea": "korea, dem. people's rep.",
        "syria":"syrian arab republic", 
        "venezuela":"venezuela, rb",
        "micronesia":"micronesia, fed. sts."
    }
    
    if element in country_reformat:
        return country_reformat[element]
    return element
```

```python
def assign_WRI_score(wri, df):
    for row in wri.itertuples():
        #row[4] is country in WRI DataFrame
        wri_country = row[4]
        wri_country_formatted = format_element(wri_country)
        for row2 in df.itertuples():
            #row2[3] is country in GDP DataFrame
            df_country = row2[3]
            if wri_country_formatted == df_country:
                #row[7] is WRI Score in WRI DataFrame
                score = row[7]
                print(score)
                row_number = df[df['Country'] == df_country].index[0]
                print(row_number)
                df.loc[row_number, 'WRI_Score'] = score
                break 
```
