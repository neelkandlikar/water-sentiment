```python
import pandas as pd
import re
import csv
from IPython.display import display, clear_output
import geopandas
```

```python
def match_aquifer(text, path_to_aquifer_file):
    tokens = text.split()
    matches = []
    with open(path_to_aquifer_file) as in_file:
        for aquifer in in_file:
            if aquifer.lower() in tokens:
                matches.append(aquifer.lower())
    return matches 
            
        
```

```python
df = pd.read_csv('../results/abstracts_cleaned_tokenized_geo_new_oct_5.csv')
display(df.head())
```

```python
df['aquifer_names_richey'] = df['text'].apply(lambda x: match_aquifer(x, '../modified_paper_aquifers.txt') )
```
