```python
import pandas as pd
import re
import csv
from IPython.display import display, clear_output
import geopandas
from ast import literal_eval
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
def format_element(element):
    if(element == 'united kingdom of great britain and northern ireland'):
        element = 'united kingdom'
    elif(element == 'america'):
        element = 'united states of america'
    return element
```

```python
df = pd.read_csv("../dataframes/12_2_capital_algorithms.csv")
```

```python
display(df.head())
```

```python
df = df.dropna(subset=['countries'])
```

```python
display(df.head())
```

```python
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
world.head()
```

```python
world['name'] = world['name'].str.lower()
world.head()
```

```python
matched = []
world['occurences'] = 0
print(type(df.loc[2, 'countries']))
for row in df.itertuples():
    countries = row[8]
   
    for element in countries:
        formatted_element = format_element(element) 
        if world['name'].str.contains(formatted_element).any() :
                row_number = world[world['name'] == formatted_element].index[0]
                world.loc[row_number, 'occurences'] +=1
#                 print(formatted_element)
#                 print('*****')
        if world['name'].str.contains(formatted_element).any() and formatted_element not in matched:
            matched.append(formatted_element)
print((matched))
```

```python
not_matched = []
all_names = world['name'].tolist()
for element in all_names:
    if(element not in matched):
        not_matched.append(element)
print(not_matched)
```

```python
display(world.head())
```

```python
world['occurences'].replace(0, np.nan, inplace=True)
world.head()
```

```python
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 50}

matplotlib.rc('font', **font)
```

```python
fig, ax = plt.subplots(1, figsize=(75,50))
world.plot(column='occurences', 
           cmap='plasma',
           alpha = 1,
           ax=ax,
           linewidth=1,
           edgecolor="red",
           legend=True,
           legend_kwds={'label': "Number of Mentions by Country",
                        'orientation': "horizontal" },
           missing_kwds={
            "color": "grey",
            "edgecolor": "red",
            "hatch": "///",
            "label": "Missing values",
        }
           
          )
```

```python
fig.savefig("../results/map_export_world.png", dpi=300)
```
