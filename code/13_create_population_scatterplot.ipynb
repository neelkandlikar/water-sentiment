```python
import pandas as pd
import plotly.express as px
from IPython.display import display, clear_output
```

```python
df = pd.read_csv("../dataframes/12_15_world_map_df.csv")
```

```python
df.head()
```

```python
df.rename(columns={'occurences':'Abstracts', 'pop_est':'Population'}, inplace=True)
```

```python
df['Country'] = df['name'].str.title()
```

```python
fig = px.scatter(df, x="Population", y="Abstracts", hover_name = "Country", trendline="ols", log_x=True, log_y=True)
fig.show()
```

```python
fig.write_html("../plots/abstracts_vs_population_log.html")
```
